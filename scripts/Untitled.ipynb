```python
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
df = pd.read_csv("/Users/jiamingqu/Desktop/trec2019/News2019/NEWS_SUBMIT/sils_news_run1.txt", sep=" ", header = None)
```

```python
df.columns = ["topicid", "q0", "newsid", "rank", "score", "runname"]
```

```python
# load qrels
qrel = pd.read_csv("../query_topics/2019BL_answer.txt", sep = " ", header = None)
qrel.columns = ["topicid", "q0", "newsid", "rel"]
```

```python
# load <topicid, newsid> pair from the xml file
tree = ET.parse('../query_topics/2019BL_topic.xml')
root = tree.getroot()
# topic to news
t2n = {}
for i in root:
    topicid = i[0].text[-3:]
    newsid = i[1].text
    t2n[topicid] = newsid
```

```python
df
```

```python
recall = 0
no_zero_rel = 0

for k in t2n.keys():
    
    print(k)
    n_rel_16 = qrel.loc[qrel.topicid == int(k)].loc[qrel.rel==16].shape[0]
    n_rel_8 = qrel.loc[qrel.topicid == int(k)].loc[qrel.rel==8].shape[0]
    n_rel_4 = qrel.loc[qrel.topicid == int(k)].loc[qrel.rel==4].shape[0]
    print("Total number of news with rel score of 16&8&4 is: {}".format(n_rel_16+n_rel_8+n_rel_4))
    rel_news_list = list(qrel.loc[qrel.topicid == int(k)].loc[qrel.rel==16].newsid) + \
                    list(qrel.loc[qrel.topicid == int(k)].loc[qrel.rel==8].newsid) + \
                    list(qrel.loc[qrel.topicid == int(k)].loc[qrel.rel==4].newsid)
    
    
    retrieved_results = list(df.loc[df.topicid == int(k)].newsid)
    rel_news_in_result = len(set(rel_news_list) & set(retrieved_results))
    print("Total number of news with rel score of 16&8&4 in cluster is: {}".format(rel_news_in_result))

    # calculate recall
    if (n_rel_16+n_rel_8+n_rel_4) == 0:
        recall_topic = 0
    else:
        recall_topic = rel_news_in_result/(n_rel_16+n_rel_8+n_rel_4)
        recall += recall_topic
        no_zero_rel += 1
    print(recall_topic)

print(recall/no_zero_rel)
```
