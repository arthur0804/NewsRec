```python
# This script calculates the measurement of distance using qrels.
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
import json
import numpy as np


def get_array(newsid):

    prob = np.zeros(80)
    file_path = "/pine/scr/j/i/jiaming/news_rec/doc_topic_prob/" + newsid + ".txt"
    try:
        with open(file_path, 'r') as f:
            for line in f.readlines():
                a = line.strip().strip('()').split(',')
                loc = int(a[0]) - 1
                val = float(a[1])
                prob[loc] = val
            return prob
    except Exception:
        return prob

def calculate_total_variation(array1, array2):
    return sum(abs(array1 - array2)) / 2




# load <topicid, newsid> pair from the xml file
tree = ET.parse('/pine/scr/j/i/jiaming/news_rec/query_topics/2018BL_topic.xml')
root = tree.getroot()

# topic to news
t2n = {}

for i in root:
    topicid = i[0].text[-3:]
    newsid = i[1].text
    t2n[topicid] = newsid


# load qrels
qrel = pd.read_csv("/pine/scr/j/i/jiaming/news_rec/query_topics/2018BL_answer.txt", sep = " ", header = None)
qrel.columns = ["topicid", "q0", "newsid", "rel"]


total_variation_16 = 0
total_variation_8 = 0
total_variation_4 = 0
total_variation_2 = 0
total_variation_0 = 0

for index, rows in qrel.iterrows():

    topicid = str(qrel.loc[index, "topicid"])
    newsid = qrel.loc[index, "newsid"]
    rel = qrel.loc[index, "rel"]

    if rel == 16:
        topic_news_array = get_array(t2n[topicid])
        candiate_news_array = get_array(newsid)
        total_variation_16 += calculate_total_variation(topic_news_array, candiate_news_array)
    elif rel == 8:
        topic_news_array = get_array(t2n[topicid])
        candiate_news_array = get_array(newsid)
        total_variation_8 += calculate_total_variation(topic_news_array, candiate_news_array)
    elif rel == 4:
        topic_news_array = get_array(t2n[topicid])
        candiate_news_array = get_array(newsid)
        total_variation_4 += calculate_total_variation(topic_news_array, candiate_news_array)
    elif rel == 2:
        topic_news_array = get_array(t2n[topicid])
        candiate_news_array = get_array(newsid)
        total_variation_2 += calculate_total_variation(topic_news_array, candiate_news_array)
    elif rel == 0:
        topic_news_array = get_array(t2n[topicid])
        candiate_news_array = get_array(newsid)
        total_variation_0 += calculate_total_variation(topic_news_array, candiate_news_array)

print(total_variation_16)
print(total_variation_8)
print(total_variation_4)
print(total_variation_2)
print(total_variation_0)
```
